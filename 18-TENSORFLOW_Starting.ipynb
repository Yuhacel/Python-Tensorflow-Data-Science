# Split Data to Test/Train

```python
from sklearn.model_selection import train_test_split
```

```python
## train_test_split
```

```python
dataFrameExcel_Bicycle["Fiyat"] ## pandas 
```

```python
## y = wx + b
## y -> label
## x -> feature
```

```python
y = dataFrameExcel_Bicycle["Fiyat"].values ## numpy array
x = dataFrameExcel_Bicycle[["BisikletOzellik1","BisikletOzellik2"]].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=15)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() 
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# TENSORFLOW


```python
import tensorflow as tf

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(4,activation="relu"))  ## 5(five) neurons
model.add(Dense(4,activation="relu"))  ## "relu" = "rectified linear unit"
model.add(Dense(4,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="rmsprop", loss="mse")  ## MSE regression 
```

```python
type(model)
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss = model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLoss = model.evaluate(x_train,y_train,verbose=0)
trainLoss
```

```python
testLoss = model.evaluate(x_test,y_test,verbose=0)
testLoss
```

```python
testGuesses = model.predict(x_test)
```
