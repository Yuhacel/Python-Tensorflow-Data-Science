```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ageList  = [10,20,30,32,34,42,45,62,64,75]
kiloList = [40,50,52,56,60,63,68,82,90,90]
```

```python
numpyAgeList = np.array(ageList)
numpyKiloList = np.array(kiloList)
```

```python
plt.plot(numpyAgeList,numpyKiloList,"g") ## "g" for "green"
```

```python
plt.plot(numpyAgeList,numpyKiloList,"r") ## "r" for "red"
```

```python
plt.plot(numpyKiloList,numpyAgeList,"b")
```

```python
plt.plot(numpyAgeList,numpyKiloList,"g")
plt.xlabel("Age")
plt.ylabel("Kilo")
plt.title("Age VS Kilo")
```

```python
numpySeries1 = np.linspace(0,10,20)
numpySeries1
```

```python
numpySeries2 = numpySeries1 ** 3
numpySeries2
```

```python
plt.plot(numpySeries1,numpySeries2,"g")
```

```python
plt.plot(numpySeries1,numpySeries2,"g--")
```

```python
plt.plot(numpySeries1,numpySeries2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpySeries1,numpySeries2,"r--")
plt.subplot(1,2,2)
plt.plot(numpySeries2,numpySeries1,"g*-")
```

```python
myFigure = plt.figure()
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpySeries1,numpySeries2,"g")
figureAxes.set_xlabel("x data name")
figureAxes.set_ylabel("y data name")
figureAxes.set_title("title here")
```

```python
newFigure = plt.figure()

axes1 = newFigure.add_axes([0.1,0.1,0.9,0.9])
axes2 = newFigure.add_axes([0.3,0.5,0.3,0.3])

axes1.plot(numpySeries1,numpySeries2,"g") ## axes1 settings
axes1.set_xlabel("x data1")
axes1.set_ylabel("y data1")
axes1.set_title("data title1")

axes2.plot(numpySeries2,numpySeries1,"b") ## axes2 settings
axes2.set_xlabel("x data2")
axes2.set_ylabel("y data2")
axes2.set_title("data title2")
```

```python
( myNewFigure, myNewAxes ) = plt.subplots(nrows=1,ncols=2)

for axes in myNewAxes:
    axes.plot(numpySeries1, numpySeries2, "g")
    axes.set_xlabel("x data")
    axes.set_ylabel("y data")

plt.tight_layout()
```

```python
newFigur = plt.figure()
## plt.figure ( dpi = 100 , figsize(3,3) )

realNewAxes = newFigur.add_axes([0.1,0.1,0.9,0.9])

realNewAxes.plot(numpySeries1, numpySeries1 ** 2, label = "numpySeries ** 2")
realNewAxes.plot(numpySeries1, numpySeries1 ** 3 , label = "numpySeries ** 3")

realNewAxes.legend(loc=2)
```
